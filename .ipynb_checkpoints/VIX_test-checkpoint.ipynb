## Import packages

```python
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

## Read VIX data

```python
vix = pd.read_csv("data/VIX_History.csv")
```

Convert into datetime and numeric

```python
vix["DATE"] = pd.to_datetime(vix.DATE)
```

```python
for col in vix.columns:
    if col != "DATE":
        vix[col] = pd.to_numeric(vix[col], downcast="float")
```

```python
import matplotlib.dates as mdates
```

```python
vix.columns
```

```python
vix["DAILY_RETURNS"] = np.log(vix.CLOSE)- np.log(vix.CLOSE.shift(1))
```

```python
fig, axs = plt.subplots(2, 1, figsize=(12, 5), sharex=True)
sns.lineplot(vix, x="DATE", y="CLOSE", ax=axs[0])
sns.lineplot(vix, x="DATE", y="DAILY_RETURNS", ax=axs[1])
axs[1].xaxis.set_major_locator(mdates.YearLocator(1, month=7, day=15))
# Remove first and last tick to prevent data outside of scope and 2024 from appearing
axs[1].set_xticks(axs[1].get_xticks()[2:-1], axs[1].get_xticklabels()[2:-1], rotation=45, ha='center')
axs[0].tick_params(bottom=False)
axs[0].set_title("VIX daily close")
axs[1].set_title("VIX daily returns");
```

## Technical indicators

```python
# Set window size
window_size = 30
```

### Moving averages

```python
vix["MOVING_AVG"] = vix.CLOSE.rolling(window_size).mean()
```

```python
fig, ax = plt.subplots(figsize=(12, 5))
sns.lineplot(vix, x="DATE", y="CLOSE", ax=ax, label="Daily close", color="lightcoral")
sns.lineplot(vix, x="DATE", y="MOVING_AVG", ax=ax, label=f"{window_size} day moving average", color="deepskyblue")
ax.xaxis.set_major_locator(mdates.YearLocator(1, month=7, day=15))
# Remove first and last tick to prevent data outside of scope and 2024 from appearing
ax.set_xticks(ax.get_xticks()[2:-1], ax.get_xticklabels()[2:-1], rotation=45, ha='center')
ax.set_title("VIX and moving average")
ax.legend();
```

### Exponential moving average

```python
vix["EWMA"] = vix.CLOSE.ewm(span=window_size).mean()
```

```python
#%matplotlib widget
fig, ax = plt.subplots(figsize=(12, 5))
sns.lineplot(vix, x="DATE", y="CLOSE", ax=ax, label="Daily close", color="lightcoral")
sns.lineplot(vix, x="DATE", y="EWMA", ax=ax, label=f"{window_size} day EWMA", color="deepskyblue")
ax.xaxis.set_major_locator(mdates.YearLocator(1, month=7, day=15))
# Remove first and last tick to prevent data outside of scope and 2024 from appearing
ax.set_xticks(ax.get_xticks()[2:-1], ax.get_xticklabels()[2:-1], rotation=45, ha='center')
ax.set_title("VIX and EWMA")
ax.legend();
```

### Momentum
NOTE: Needs further research into this indicator

```python
# N day momentum is defined as P_i - P_i-5
vix["MOMENTUM"] = vix.CLOSE- vix.CLOSE.shift(window_size)
```

```python
fig, ax = plt.subplots(figsize=(12, 5))
sns.lineplot(vix, x="DATE", y="CLOSE", ax=ax, label="Daily close", color="lightcoral")
sns.lineplot(vix, x="DATE", y="MOMENTUM", ax=ax, label=f"{window_size} day momentum", color="deepskyblue")
ax.xaxis.set_major_locator(mdates.YearLocator(1, month=7, day=15))
# Remove first and last tick to prevent data outside of scope and 2024 from appearing
ax.set_xticks(ax.get_xticks()[2:-1], ax.get_xticklabels()[2:-1], rotation=45, ha='center')
ax.set_title("VIX and momentum")
ax.legend();
```

## Autocorrelation and partial autocorrelation
NOTE: Install package `statsmodels` for plotting acf and pcf

```python
pd.plotting.autocorrelation_plot(vix.CLOSE)
```

```python
# Plot 30 day autocorrelation
num_lags = 20
autocorrs = np.zeros(num_lags)
for i in range(num_lags):
    autocorr = vix.CLOSE.autocorr(lag=i+1)
    autocorrs[i] = autocorr
ax = sns.scatterplot(x=np.arange(1, num_lags+1), y=autocorrs, s=70, color="royalblue")
ax.axes.axhline(color="black")
for i, a in enumerate(autocorrs):
    ax.plot([i+1, i+1], [0,a], color="royalblue")
ax.set_xlabel("Number of lags")
ax.set_ylabel("Autocorrelation");
```

```python
from statsmodels.graphics.tsaplots import plot_pacf, plot_acf
```

```python
plot_pacf(vix.CLOSE, lags=20, method="ywm")
plot_acf(vix.CLOSE, lags=20);
```
