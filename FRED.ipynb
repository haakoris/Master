Monthly unemployment rate (UNRATE)

```python
UNRATE = fred.get_series('UNRATE', observation_start='1990-01-02')
UNRATE = pd.DataFrame(UNRATE, columns=["UNRATE"])
UNRATE.index = pd.to_datetime(UNRATE.index, format='%Y-%m-%D')
start_date = UNRATE.index.min() - pd.DateOffset(day=1)
end_date = UNRATE.index.max() + pd.DateOffset(day=31)
dates = pd.date_range(start_date, end_date, freq='D')
UNRATE = UNRATE.reindex(dates, method='ffill')
UNRATE = UNRATE.tail(-1)
frames_vix_dataset.append(UNRATE)
UNRATE.head(32)
```
